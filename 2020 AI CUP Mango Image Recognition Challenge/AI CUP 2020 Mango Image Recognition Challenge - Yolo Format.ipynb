```python
import pandas as pd
```

```python
train = pd.read_csv('train1.csv', header=None)
train.head(5)
```

```python
for i in range(len(train)):
    print("%s %s %s %s %s" %(train[5][i] ,train[1][i], train[2][i], train[3][i], train[4][i]), 
          file=open('Txt/%s.txt' %(train[0][i][:-4]), 'w'))
    k = 6
    while pd.isna(train[6][i])==False:
        print("%s %s %s %s %s" %(train[k+4][i], train[k][i], train[k+1][i], train[k+2][i], train[k+3][i]), 
              file=open('Txt/%s.txt' %(train[0][i][:-4]), 'a'))
        k = k+5
        if k >=train.shape[1]:
            break
        if pd.isna(train[k][i]):
            break
```
