```python
import tkinter as tk 
import pandas as pd 
import numpy as np
from tkinter import ttk
from tkinter import messagebox 
from tkinter.font import Font 
from PIL import ImageTk, Image 
from tkinter import messagebox 
from tkinter.filedialog import askopenfilename
from tensorflow.keras.models import load_model
import warnings
warnings.filterwarnings("ignore")

feature_name = ['Good', 'Bad','None']
```

```python
#建立選擇欄
def NewFile():
    print("New File!")
def About():
    messagebox.showinfo(title='About', message="本研究東華大學測試使用!")
```

```python
def open_model_Resnet():
    filename = askopenfilename(title='Select file', filetypes=[("all files","*.*"),("keras files","*.h5"),("tensorflow files","*.tf"),("darknet files","*.weight")])
    modelpath_Var[0].set(filename)
    
def open_model_Inception():
    filename = askopenfilename(title='Select file', filetypes=[("all files","*.*"),("keras files","*.h5"),("tensorflow files","*.tf"),("darknet files","*.weight")])
    modelpath_Var[1].set(filename)
    
def open_model_DenseNet():
    filename = askopenfilename(title='Select file', filetypes=[("all files","*.*"),("keras files","*.h5"),("tensorflow files","*.tf"),("darknet files","*.weight")])
    modelpath_Var[2].set(filename)
    
def open_model_SVM():
    filename = askopenfilename(title='Select file', filetypes=[("all files","*.*"),("keras files","*.h5"),("tensorflow files","*.tf"),("darknet files","*.weight")])
    modelpath_Var[3].set(filename)
```

```python
def clear_text():
    for i in range(len(entry_L)):
        entry_L[i].delete(0, 'end')
```

```python
def open_img():
    
    filename = askopenfilename(title='Select file', 
                                          filetypes=[("all files","*.*"),
                                                     ("png files","*.png"),
                                                     ("jpg files","*.jpg"),
                                                     ("jpeg files","*.jpeg")])
    img = Image.open(filename).resize((295,240))
    imgtk = ImageTk.PhotoImage(img)
    label_img.config(image = imgtk)
    label_img.image = imgtk
    
    img_var.set(filename)
```

```python
def predict():
    img_predict = Image.open(str(img_var.get())).resize((224,224))
    npimg = np.asarray(img_predict, dtype="uint8")
    img_predict_nor = npimg.astype('float32')/255
    
    th_0, th_1, th_3, th_4 = 0.5, 0.5, 0.5, 0.5
    
    #=================================
    if modelpath_Var[0].get()!= "":
        model = load_model(modelpath_Var[0].get(), compile = False)
        predict_0 = model.predict(np.array([img_predict_nor]))
        
        if entry_L[0].get()!= "":
            th_0 = entry_L[0].get()
            
        if predict_0[0][1] > float(th_0):
            predictions_0 = [1]
        else:
            predictions_0 = [0]
        
    else:
        predict_0 = [[np.nan, np.nan, np.nan]]
        predictions_0 = [2]
    
    #=================================
    if modelpath_Var[1].get()!= "":
        model = load_model(modelpath_Var[1].get(), compile = False)
        predict_1 = model.predict(np.array([img_predict_nor]))
        predictions_1 = [np.argmax(value)for value in predict_1]
        
        if entry_L[1].get()!= "":
            th_1 = entry_L[1].get()
            
        if predict_1[0][1] > float(th_1):
            predictions_1 = [1]
        else:
            predictions_1 = [0]
            
    else:
        predict_1 = [[np.nan, np.nan, np.nan]]
        predictions_1 = [2]
    
    #=================================
    if modelpath_Var[2].get()!= "":
        model = load_model(modelpath_Var[2].get(), compile = False)
        predict_2 = model.predict(np.array([img_predict_nor]))
        predictions_2 = [np.argmax(value)for value in predict_2]

        if entry_L[2].get()!= "":
            th_2 = entry_L[2].get()
        
        if predict_2[0][1] > float(th_2):
            predictions_2 = [1]
        else:
            predictions_2 = [0]
            
    else:
        predict_2 = [[np.nan, np.nan, np.nan]]
        predictions_2 = [2]
    
    #=================================
    if modelpath_Var[3].get()!= "":
        model = load_model(modelpath_Var[3].get(), compile = False)
        predict_3 = model.predict(np.array([img_predict_nor]))
        predictions_3 = [np.argmax(value)for value in predict_3]
        
        if entry_L[3].get()!= "":
            th_3 = entry_L[3].get()
        
        if predict_3[0][1] > float(th_3):
            predictions_3 = [1]
        else:
            predictions_3 = [0]
        
    else:
        predict_3 = [[np.nan, np.nan, np.nan]]
        predictions_3 = [2]
    
    messagebox.showinfo(title='Prediction', message=
    "ResNet \n Prediction: %s \n Probability: %0.4f \n\n" %(feature_name[predictions_0[0]], predict_0[0][predictions_0[0]])+
    "Inception \n Prediction: %s \n Probability: %0.4f \n\n" %(feature_name[predictions_1[0]], predict_1[0][predictions_1[0]])+
    "DenseNet \n Prediction: %s \n Probability: %0.4f \n\n" %(feature_name[predictions_2[0]], predict_2[0][predictions_2[0]])+
    "SVM \n Prediction: %s \n Probability: %0.4f" %(feature_name[predictions_3[0]], predict_3[0][predictions_3[0]]))
```

```python
#建立視窗
app = tk.Tk()

#建立視窗標題
app.title('Soybeans - Defect detection')

#建立視窗背景
#app.configure(background = 'white')

#建立視窗大小
app.geometry('380x640')

#=================================================================
#建立菜單
menu = tk.Menu(app)
app.config(menu=menu)

#建立菜單的file
file_menu = tk.Menu(menu, tearoff = 0) #是否有虛線，無虛線則無法分離視窗
menu.add_cascade(label="File", menu=file_menu)

#file建立新增
file_menu.add_command(label="New", command=NewFile)

#file建立開啟
file_menu.add_command(label="Open Image", command=open_img)

#file建立虛線
file_menu.add_separator()

#file建立離開
file_menu.add_command(label="Exit", command=app.destroy)

#建立菜單的幫助
help_menu = tk.Menu(menu, tearoff = 0)
menu.add_cascade(label="Help", menu=help_menu)
help_menu.add_command(label="About", command=About)

#=================================================================
#設立字體
title_label_Font = Font(family="Times", size=18, underline=1)
subtitle_label_Font_1 = Font(family="Times", size=14)

#建立畫布與位置
frame_title = tk.Frame(app)
frame_title.grid(column=0, row=0, ipadx=5, pady=5)

frame_input = tk.LabelFrame(app, text='Model')
frame_input.grid(column=0, row=1, ipadx=5, pady=5)

frame_img = tk.LabelFrame(app, text='Image show')
frame_img.grid(column=0, row=2, ipadx=5, pady=5)

frame_output = tk.Frame(app)
frame_output.grid(column=0, row=3, ipadx=5, pady=5)

#=================================================================
#建立標題與位置
label_Title = tk.Label(frame_title,text="Soybeans - Defect detection", font=title_label_Font)
label_Title.grid(column=0, row=0, ipadx=5, pady=5)

label_Subtitle = tk.Label(frame_title,text="Please enter the threshold and choose to open the model.", font=subtitle_label_Font_1)
label_Subtitle.grid(column=0, row=1, ipadx=5, pady=5)

#=================================================================
#模型名稱與輸入欄的放置位置(左與右)
label_name_L = ['ResNet', 'Inception', 'DenseNet', 'SVM']
fun_command = [open_model_Resnet, open_model_Inception, open_model_DenseNet, open_model_SVM]

entry_L, button_model = [], []

for i in range(len(label_name_L)):
    label = tk.Label(frame_input,text = label_name_L[i])
    label.grid(column=0, row=2+i, ipadx=5, pady=5, sticky=tk.W+tk.N)
    
    entry_L.append(tk.Entry(frame_input, width=10))
    entry_L[i].grid(column=1, row=2+i, padx=10, pady=5, sticky=tk.N)
    
    button_model.append(tk.Button(frame_input, text = 'Open file',command = fun_command[i]))
    button_model[i].grid(column=2, row=2+i, padx=10, pady=5, sticky=tk.N)

#=================================================================
#放置變數
modelpath_Var = []
for i in range(len(label_name_L)):
    var = tk.StringVar()
    modelpath_Var.append(var)
    
img_var = tk.StringVar()
    
#=================================================================
#影像放置位置

img = Image.open("pre-inserted picture.png").resize((295,240))
imgtk = ImageTk.PhotoImage(img)
label_img = tk.Label(frame_img, image=imgtk)
label_img.grid(column=0, row=0, ipadx=5)

openimgButton = tk.Button(frame_img, text = 'Open Image',command = open_img)
openimgButton.grid(column=0, row=1, pady=5)

#=================================================================
#建立按鈕與位置
resultButton = tk.Button(frame_output, text = 'Get Result',command = predict)
resultButton.grid(column=0, row=0, pady=10)

clear_button = tk.Button(frame_output, text=" Clear text", command = clear_text)
clear_button.grid(column=1, row=0, pady=10)

#程式開始迴圈
app.mainloop()
```
